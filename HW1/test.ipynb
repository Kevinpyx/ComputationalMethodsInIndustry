### Tests for Bisection method


```python
import root_finding
import math

def func0(x): 
    return (2*x-1)

def func1(x):
    return x**3 - 3*x + 1
# a: find the root in [1, 2]
# b: find the root in [0, 1]

def func2a(x):
    return x**2 - 2

def func2b(x):
    return x**2 - 3

def func3(x):
    return math.cos(x)-x
```

```python
# test 0
print("result for test 0")
print(root_finding.bisect(func0, -2, 4))
```

```python
# test 1
print("result for test 1")
print(root_finding.bisect(func1, 1, 2))
```

```python
# test 1b
print("result for test 1b")
print(root_finding.bisect(func1, 0, 1))
```

```python
# test 2a
print("result for test 2a")
print(root_finding.bisect(func2a, 1, 2))
```

```python
# test 2b
print("result for test 2b")
print(root_finding.bisect(func2b, 1, 2))
```

```python
# test 3
print("result for test 3")
print(root_finding.bisect(func3, 0, 2))
```

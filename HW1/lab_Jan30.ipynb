```python
import root_finding as rf
import math
```

```python
#4
def func4(x):
    return x**3 - 3*x + 1

def dfunc4(x):
    return 3*x**2 - 3

print(rf.newton(func4, dfunc4, -0.8))
```

Interpretation of the problem:
The tangent line at -0.8 led us to the right side of root 1.532. Therefore, the function just found root 1.532, rather than the root between 0 and 1. We can't find the desired root starting with -0.8.

```python
#5
def func5(x):
    return x**3 - 2*x + 2

def dfunc5(x):
    return 3*x**2 - 2

print(rf.newton(func5, dfunc5, 0))
```

Interpretation of the problem:
The tangent line at 0 leads us to root x_1=1. The tangent line at 1 leads us back to 0. So x oscillates between 0 and 1 and never converges. We can't find the desired root starting with 0.

```python
#6 starting from -3
def func6(x):
    return math.exp(x)/(1+math.exp(x)) - 0.5

def dfunc6(x):
    return math.exp(x)/((1 + math.exp(x))**2)

print(rf.newton(func6, dfunc6, -3))
```

```python
#6 starting from -2
print(rf.newton(func6, dfunc6, -2))
```

Interpretation of the problem:
The newton's method doesn't work when x_0=-3 but works when x_0=-2. The problem with -3 is the gradient at -3 is too flat so it leads us to a point with greater x (absolute) value, where the gradient is even flatter. And then we get to a point even farther away from the root and lastly there is a math range error because the number goes out of the range. 
-2 works because the gradient is steep enough for us to converge to 0. 

```python
#7
def func7(x):
    return math.sin(x) - x

def dfunc7(x):
    return math.cos(x) - 1

print(rf.newton(func7, dfunc7, -2*math.pi))
```

Interpretation of the problem:
The Newton method works for most x values except integer multiples of 2Ï€, where the slope becomes 0.

```python
#8a
def func8a(x):
    return abs(x)**(1/3)

def dfunc8a(x):
    #print("evaluating derivative:", x)

    if x<0:
        # 1/3 * (-x)^(-2/3)
        return -((x**2)**(-1/3))/3
    if x>0:
        return (1/3)*(x**(-2/3))
    else:
        raise Exception("Attempted to evaluate derivative at 0")

print(rf.newton(func8a, dfunc8a, 1))
```

Interpretation of the problem:
The gradient is too flat so the value diverges.

```python
#8b
def func8b(x):
    return abs(x)**(2/3)

def dfunc8b(x):
    #print("evaluating derivative:", x)

    if x<0:
        return 2/3*(-x)**(-1/3)
    if x>0:
        return (2/3)*(x**(-1/3))
    else:
        raise Exception("Attempted to evaluate derivative at 0")
    
print(rf.newton(func8b, dfunc8b, 1))
```

```python
#8c
def func8c(x):
    return abs(x)**(4/3)

def dfunc8c(x):
    #print("evaluating derivative:", x)

    if x<0:
        return 4/3*(-x)**(1/3)
    if x>0:
        return (4/3)*(x**(1/3))
    else:
        raise Exception("Attempted to evaluate derivative at 0")
    
print(rf.newton(func8c, dfunc8c, 1))
```

Interpretation of the problem:
Only the last one converges because the numerator is smaller than the denominator. The others don't converge at all no matter how close from 0 we start.
